```python
import random as r
from math import sqrt
```

```python
def distance(a, b):
    return sqrt(a**2+b**2)
```

```python
def estimate_pi(darts):
    inside = 0
    
    for i in range(darts):
        x = r.random()
        y = r.random()
        
        if distance(x,y) < 1.0:
            inside += 1
    pi = (inside/darts) * 4
    return pi
```

```python
estimate_pi(int(1e+7))
```

```python
import random

def get_action():
    return random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
```

```python
def random_walk(n):
    path = []
    
    x, y = 0, 0
    for i in range(n):
        dx, dy = get_action()
        
        x = x + dx
        y = y + dy
        
        path.append((x,y))
        
    return path
```

```python
walk = random_walk(2000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
steps = 2000
plt.figure(num= None,figsize=(15,10))
point_color = list(range(steps))
```

```python
x_numbers = [x[0] for x in walk]
y_numbers = [x[1] for x in walk]
```

```python
plt.scatter(x_numbers,y_numbers,marker=".",s=100,c=point_color)
plt.show()
```
